```python
import pandas as pd

#create dataset

data = {
    "product": ["apple", "banana", "orange", "apple", "banana", "orange"],
    "region": ["north", "north", "south", "south", "north", "south"],
    "sales": [120, 80, 150, 90, 200, 50]
}

#load into dataframe
df = pd.DataFrame(data) 

#import matplot library
import matplotlib.pyplot as plt

#group by region and calculate total sales
region_sales = df.groupby("region")["sales"].sum().reset_index()
print(region_sales)

#create bar graph by product x sales
region_sales.plot(kind='bar', x='region', y='sales', title='Total Sales per region')
plt.ylabel('total sales') # Add y-axis label
plt.show()
```

```python
import pandas as pd
import numpy as np

#create dataset

data = {
    "product": ["apple", "banana", "orange", "apple", "banana", "orange"],
    "region": ["north", "north", "south", "south", "north", "south"],
    "sales": [120, 80, 150, 90, 200, 50]
}

#load into dataframe
df = pd.DataFrame(data) 

#import matplot library
import matplotlib.pyplot as plt
%matplotlib inline

# Add a new column "sales_category" indicating high or low
df['sales_category'] = np.where(df['sales'] > 100, 'high', 'low')
print(df)

#group by sales_category
grouped_data = df.groupby('sales_category')["sales"].sum()

#create bar graph by product x sales
grouped_data.plot(kind='pie', autopct='%1.1f%%' ,title='Sales by Category')
plt.ylabel('') # Add y-axis label
plt.show()
```
