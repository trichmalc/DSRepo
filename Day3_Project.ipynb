```python
#importing pandas

import pandas as pd

#create dataset

data = {
    "product": ["apple", "banana", "orange", "apple", "banana", "orange"],
    "region": ["north", "north", "south", "south", "north", "south"],
    "sales": [120, 80, 150, 90, 200, 50]
}

df = pd.DataFrame(data) 

# Filter for rows where sales is greater threshold of 100
filtered_df = df[df["sales"] > 100]
print(filtered_df)

# Filter for rows where sales is greater than 100 AND region is 'north'
filtered_df_multiple = df[(df["sales"] > 100) & (df["region"] == 'north')]
print(filtered_df_multiple)

# Filter for rows where sales is greater than 100 AND region is 'north' using query
filtered_query = df.query('sales > 100 and region == "north"')
print(filtered_query)
```

```python
#importing pandas
import pandas as pd

#creating dataset
data2 = {
    "product": ["apple", "apple", "orange", "apple", "orange", "orange"],
    "date": ["Aug 20", "Aug 21", "Aug 22", "Aug 23", "Aug 24", "Aug 25"],
    "sales": [12, 8, 15, 9, 20, 5]
}

df = pd.DataFrame(data2)
df

#import matplot library
import matplotlib.pyplot as plt

#create bar graph by product x sales
df.plot(kind='bar', x='product', y='sales', title='Sales per product')
plt.ylabel('sales') # Add y-axis label
plt.show()

# Calculate the cumulative sum of the 'Value' column
df['Cumulative_Sum'] = df['sales'].cumsum()
print("\nDataFrame with cumulative sum for 'sales':")
print(df)

 # Using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Cumulative_Sum'], marker='o')
plt.title('Cumulative Sum Line Plot')
plt.xlabel('Day')
plt.ylabel('Cumlative Sales')
plt.show()

```
